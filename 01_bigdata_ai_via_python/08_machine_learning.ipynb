### 기계학습 모델(Machine Learning Models)
  * 질적(qualitative) 값, 수적/양적 값(quantitative)
    * 질적 값: 분류(classification)
    * 양적 값: 회귀(regression)
  * 손실함수(loss function): 입력값 x에 따라 결정되는 예측값 y가 실제 측정된 결과값의 y와 얼마나 떨어져 있는가 계산

#### 지도학습 모델 종류
  * 결정트리(decision tree)
  * 최근접 이웃 탐색(nearest neighbor search)
  * 서포트 벡터 머신(SVM: support vector machine)
  * 선형 회귀(linear regression)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(f'예측:\n{predictions}')
print(f'정답(y_test):\n{y_test}')
```

#### 최근접 이웃탐색(Nearest Neighbor Search) 모델
##### K-최근접 이웃(k-NN: k-NearestNeighbors) 모델
  * 분류나 회귀에 사용되는 방식
  * 입력이 특정 공간 내에서 예측하고자 하는 대상과 가장 가까운 k개의 대상을 찾도록 훈련하는 방식
  * 출력은 k-NN이 분류로 사용되었는지 또는 회귀로 사용되었는지에 따라 달라짐
    * k-NN 분류에서 출력은 소속된 항목
    * k-NN 회귀에서 출력은 객체의 특성값
