```python
from scipy.spatial import distance
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def euc(a, b) -> float:
    return distance.euclidean(a, b)


class EucKNN():
    def fit(self, X_train, y_train) -> None:
        self.X_train = X_train
        self.y_train = y_train
    
    def predict(self, X_test) -> list:
        predictions = [self.closest(row) for row in X_test]
        return predictions
    
    def closest(self, row):
        best_dist = euc(row, self.X_train[0])
        best_index = 0
        for i in range(1, len(self.X_train)):
            dist = euc(row, self.X_train[i])
            if dist < best_dist:
                best_dist = dist
                best_index = i
        return self.y_train[best_index]
```

```python
iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

clf = EucKNN()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(accuracy_score(y_test, predictions))
# 유클리드 거리를 구하는 방법을 적용한 결과 90% 이상의 높은 예측 결과를 보여줌
```

##### 서포트 벡터 머신(SVM: Support Vector Machine)
  * 지도학습 모델의 하나로, 본류와 회귀 방식 모두에 사용 가능
  * 선형적인 경우뿐만 아니라 초평면(hyper plane)과 같은 비선형적인 경우에도 사용할 수 있음
    * 패턴 인식이나 자료 분석을 위한 학습 모델에는 비선형적인 입력 데이터를 사용할 수 없음

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(y_test)

print(accuracy_score(y_test, predictions))
```

* 서포트 벡터 머신에서는 감마와 C값이 있음
    * 감마 값을 증가시키면 결정 경계가 복잡해져 과도한 연산을 초래할 수 있으나 인식률 향상이 가능
    * 데이터의 양과 특성에 따라 적절한 값을 설정해야 함
    * C값은 학습 시 잘못된 분류를 생성할 때 부여하는 벌점을 의미
    * C값을 크게 설정하면 분류 결과에 오버피팅 초래 가능성 커지고 작게 설정하면 언더피팅 초래 가능성 커짐
        * 언더피팅 - 예측률 저하시키는 원인
        * 오버피팅 - 기계학습 시 과도한 연산 초래

##### 선형 회귀(Linear Regression)
  * 선형 회귀분석은 변수들 사이의 관계를 분석하는 통계학적 방법 사용
    * 알고리즘이 비교적 간단하며 다양한 문제에 적용할 수 있다는 장점을 가짐
  * 더욱 정확한 예측을 위하여 error를 최소화하는 a와 b의 최적값을 설정해야 함
    * 이를 위해 Cost Function이 필요하며, 이는 training set을 가장 적은 오차로 표현하고 있는지 확인하는 함수
    * Loss Function(손실함수) 또는 Objective Function 이라고도 함
  * Linear Regression에서는 평균 제곱 오차(Mean Squared Error) 방식을 사용하여 오차 최소화
  * 최적의 a, b 값을 구하기 위하여 Gradient descent 방식 사용
    * 훈련 데이터의 값을 입력하여 계산된 출력값과 해당 입력값에 대한 실제 출력값의 차이를 제곱하여 평균을 구함
    * 이때 계산된 값이 0에 가까울수록 더욱 정확한 함수(직선)
  * `loss = tf.reduce_mean(tf.square(y - y_data))`
    * `y`: 입력값에 따라 계산된 예측값
    * `y_data`: 정해진 결과값

```python
# [1] 케라스 버전을 위한 데이터 준비
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = 500
vectors = []

for i in range(points):
    x1 = np.random.normal(0.0, 0.5)
    y1 = x1 * 0.3 + 0.5 + np.random.normal(0.0, 0.05)
    vectors.append([x1, y1])

x_data = [v[0] for v in vectors]
y_data = [v[1] for v in vectors]

plt.scatter(x_data, y_data, color='yellow', label='Training Data')
```

```python
# [2] 딥러닝 모델링
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential(
    [layers.Dense(units=1, input_shape=[1])]
)
model.compile(loss='mse', optimizer='SGD') 
# 'mean_squared_error'
history = model.fit(x_data, y_data, epochs=100, verbose=False)
```

```python
# [3] 경계선 그래프
weights, bias = model.get_weights()
result_line = x_data * weights[0] + bias

plt.scatter(x_data, y_data, color='yellow', label='Training Data')
plt.plot(x_data, result_line, color='red', label='Result Line')
plt.legend()
plt.show()
```

```python
# [4] 딥러닝 모델 테스트
test = np.array([[-1], [0], [1]])
pred = model.predict(test)
print(pred)
```

##### K-Means 방식
  * 입력 데이터를 K개 클러스터로 분류
  * 각 데이터들이 어느 클러스터에 속하는지 판단하기 위해 각 클러스터에 임의의 중심값 선택
  * 그리고 그 중심에서 각 데이터의 거리를 반복적으로 계산하여 각 데이터의 가장 가까운 중심을 갱신하는 방식

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 5], [3, 4], [2, 3], [5, 3], [6, 2], [5, 1]])
kmeans = KMeans(n_clusters=2, random_state=42).fit(X)

# 속성 kmenas.labels_에 생성된 레이블 출력
print(kmeans.labels_)
# 레이블 1의 중간점을 각각 출력
# 레이블 0의 중간점 출력
print(kmeans.cluster_centers_)
```

```python
y = kmeans.labels_
ycenter = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.scatter(ycenter[:, 0], ycenter[:, 1], c=[0, 1], s=200, cmap='viridis', edgecolor='k')
```

  * `corr()` 함수는 피어슨 상관계수를 이용한 것
  * 두 변수 간 선형 관계를 계산하는 방식
  * 입력된 데이터에서 칼럼(column)의 상관관계 계산

```python
# `corr()` 함수를 이용하여 상관관계 측정
corr_flavors = pd.DataFrame.corr(flavors)
print(corr_flavors)
```

```python
import matplotlib.pyplot as plt
```
