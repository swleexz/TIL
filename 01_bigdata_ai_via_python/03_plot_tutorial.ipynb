```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
# 'bo': b(blue) 색상의 "o" 모양으로 점 표시
plt.plot(x, y, 'bo')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
# 'bo': b(blue) 색상의 "o" 모양으로 점 표시
plt.plot(x, y, 'bo')
plt.axis([0, 6, 0, 30]) # x축: 0~6, y축: 0~30 범위
plt.show()
```

#### Summary: matplotlib.pyplot
* legend 추가: legend()
* 축 교정: axis()
* 축의 라벨 설정: xlabel(), ylabel()
* 그림 저장: savefig()

```python
import numpy as np
points = np.array([[1, 1], [1, 2], [1, 3], [2, 1], [2, 2], [2, 3], [3, 1], [3, 2], [3, 3]])

p = np.array([2.5, 2])
plt.plot(points[:, 0], points[:, 1], 'ro')
plt.plot(p[0], p[1], 'bo')
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 20)
y = x ** 2.0
plt.plot(x, y, 'bo-', linewidth=3, markersize=5)
```

```python
# 'gs-': green square, 이어짐
plt.plot(x, y, 'gs-', linewidth=1, markersize=3)
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2.0
y2 = x ** 1.5
plt.plot(x, y1, 'bo-', linewidth=2, markersize=12, label='First')
plt.plot(x, y2, 'gs-', linewidth=2, markersize=12, label='Second')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc='upper left')
# pdf 파일로 저장 가능
plt.savefig('mplot.pdf')
```

```python
x = np.logspace(-1, 1, 20)
y1 = x ** 2.0
y2 = x ** 1.5
plt.plot(x, y1, 'bo-', linewidth=2, markersize=5, label='First')
plt.plot(x, y2, 'gs-', linewidth=2, markersize=5, label='Second')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc='upper left')
plt.savefig('mplot.pdf')
```
